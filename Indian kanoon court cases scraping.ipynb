## Scraping in 4 parts


```python
requests2018_21 = []

for number in numbers[:39]:
    
    with driver:
        driver.get(f'https://indiankanoon.org/search/?formInput=citedby%3A%201641007%20sortby%3A%20mostrecent&pagenum={number}')
        
        time.sleep(10)

        results = driver.find_elements_by_class_name('result')

        for each in results:
            names = each.find_elements_by_class_name('result_title')
            for name in names:
                title = name.text

            courtags = each.find_elements_by_class_name('docsource')
            for court in courtags:
                courtname = court.text

            citetags = each.find_elements_by_class_name('cite_tag')
            for citetag in citetags:
                if 'doc' in citetag.get_attribute('href'):
                    doc = citetag.get_attribute('href')


            request = {
                    'case_title': title,
                    'court_name' : courtname,
                    'court_doc': doc
                }

            requests2018_21.append(request)
            
                
#             fulldoc.click()
#             driver.find_element_by_xpath('/html/body/div[5]/form[1]/input[2]').click()
#             driver.back()
                
        time.sleep(2)

requests2018_21
```
