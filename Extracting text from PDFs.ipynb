## Extracting text from clean PDFs
I obtained a file list of the locally saved PDFs using `cd` on the command line.

```python
filelist_sedition = pd.read_excel('filelist_IK_sedition_cases.xlsx')
```

```python
filelist_sedition = filelist_sedition.filelist_IndianKanoon_PDFs_Sedition.to_list()
```

```python
filelist_sedition
```

```python
import PyPDF2

textfromfiles = []
pagenum = -1 

for eachfile in filelist_sedition:
    pdfFileObj = open(f'IndianKanoonPDFs_Section124A/{eachfile}', 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    count = (pdfReader.numPages)

    numlist = list(range(count))
    textlist = []  
    
    for num in numlist:
        pageObj = pdfReader.getPage(num)
        
        try:
            text=(pageObj.extractText())
            text=text.split(",")
    #         text
        except:
            text=f'{eachfile} on page number {num+1} had an error.'
        
#         if num+1 == count:
#             textlist.append(text)
#         else:
#             text2 = text
        
        textlist.append(text)
            
        eachpdftext = {
            'pdf_name': eachfile,
            'page_number': num+1,
            'max_pages': count,
            'text': textlist
            
        }
        
        textfromfiles.append(eachpdftext)
        
textfromfiles

df = pd.DataFrame(textfromfiles)
df
```
